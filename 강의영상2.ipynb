### 변수에 값을 넣는 것 => 할당

sum = 1+2=> 1+2의 계산된 결과를 sum이라는 변수에 할당한다.

sum 변수명이고 "=" 같다는 의미가 아니라 할당(assignment)이라고 합니다.

"=" : 할당 연산자
"+" : 산술 연산자

+= : 산술 후 바로 할당하는 문법
(/=,-=,*=)모두 가능합니다.

```python
# 변수를 사용할 떄
# 이전에 계산된 결과를 저장해서 사용할수 있다
# 변수를 사용하면서 사용 가능한 문법

sum = 1+2
sum = sum+3
sum = sum+4
sum = sum+5

print(sum)

sum2 = 1+2
sum2 += 3
sum2 += 4
sum2 += 5

print(sum2)
```

```python
# 주의!! 선언되지 않은 변수를 사용할 경우 에러 발생

sum3 = 0
sum3 += 1

sum3
```

참고) 파이썬에서 정수 256 까지는 이미 해당 값이 메모리 상에 존재하면 기존의 객체를 재활용한다고 합니다.(신기방기)
1 == 1.0 을 입력했을 때에는 True
1 is 1.0 을 입력했을 때에는 False
1과 1.0은 같은 값이지만
1은 정수이고, 1.0은 소수점을 가진 실수이다.
둘의 값은 같지만 객체가 완전히 동일하지 않고 type이 다르다.
객체는 생성하면서 메모리에 저장되며 이때 고유한 ID가 생기면서 메모리 주소값을 가진다고 합니다.
따라서 값이 일치하는지 확인하려면 is 가 아닌 ==를 써야 한다고 합니다.
