__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다. 

마법을 사용하려면 지능이 50 이상이어야 합니다. 

마법사의 지능이 50 이상인지 확인하세요.

```python
직업 = "마법사"
힘 = 10
민첩 = 20
지능 = 80
체력 = 50

print(f"{직업} 지능{지능}")
print(f"마법사 지능은 50보다 높다 {50 <= 지능}")
```

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다. 


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
마법사 = 120
전사 = 150
도적 = 110
힐러 = 90

print(f"마법사{100<=마법사}")
print(f"전사{100<=전사}")
print(f"도적{100<=도적}")
print(f"힐러{100<=힐러}")
```
